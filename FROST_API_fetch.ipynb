```python
sources = pd.DataFrame()

for row in data:
    row_source = pd.DataFrame([row])
    sources = pd.concat([sources, row_source])

sources = sources.reset_index(drop=True)

sources
```

```python
sensors = sources[sources['@type'] == "SensorSystem"]
sensors_norway = sensors[sensors["countryCode"] == 'NO'].reset_index(drop=True)
```

```python
selected = sensors_norway[["id", "geometry", "county", "municipality", "validFrom", "masl"]]
selected = selected.dropna(subset=["geometry"])
selected["geometry"] = selected["geometry"].apply(lambda x: x["coordinates"])
selected = selected.reset_index(drop=True)
# selected.to_csv("norway_weather_stations.csv")
selected
```

```python
stations = []
for station in selected["id"]:
    stations.append(station)

elements_str = "mean(air_temperature P1D), mean(max(wind_speed PT1H) P1D), min(max(wind_speed PT1H) P1D), mean(wind_speed P1D), \
mean(wind_speed_of_gust P1D), sum(precipitation_amount P1D)"
time_str = "2020-12-01/2020-12-31"
print(f"number of stations: {len(stations)}, stations -> {stations}")
print(elements_str + "\n" + time_str)
```

```python
weather_data = []
errors = []

st_number = 0
for station in stations:
    # Define endpoint and parameters
    endpoint = 'https://frost.met.no/observations/v0.jsonld'
    parameters = {
        'sources': station,
        'elements': elements_str,
        'referencetime': time_str,
    }
    # Issue an HTTP GET request
    r = requests.get(endpoint, parameters, auth=(client_id,''))

    # Check if the request worked, print out any errors
    if r.status_code == 200:
        # Extract JSON data
        json = r.json()
        data = json['data']
        print(f'Data retrieved from frost.met.no for station -> {station}')
    else:
        print('Error! Returned status code %s' % r.status_code + f' for station -> {station}')
        errors.append(f"Error_{r.status_code}_Station_{station}")
        continue

    a = pd.DataFrame()

    for row in data:
        row_source = pd.DataFrame([row])
        a = pd.concat([a, row_source])

    a = a.reset_index(drop=True)

    weather_data.append(a)
    print(f"Finished data processing for station -> {station}, number -> {st_number}!")
    st_number += 1
```
