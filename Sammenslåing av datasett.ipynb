```python
import pandas as pd
import matplotlib.pyplot as ply
```

```python
ds1 = pd.read_csv("Summert_forbruk_sone.csv", delimiter=",")
ds1 = ds1.rename(columns={'VOLUM_KWH': 'FORBRUK_KWH'})

ds1.head()
```

```python
ds2 = pd.read_csv("Summert_produksjon_sone.csv", delimiter=",")
ds2 = ds2.rename(columns={'VOLUM_KWH': 'PRODUKSJON_KWH'})

ds2.head()
```

```python
ds1 = ds1.drop(columns=['Unnamed: 0'])
ds2 = ds2.drop(columns=['Unnamed: 0'])

merged_ds = pd.merge(ds1, ds2, on=['DATE', 'PRISOMRÅDE'], how='inner')
merged_ds['NETTO_PRODUKSJON_KWH'] = merged_ds['PRODUKSJON_KWH'] - merged_ds['FORBRUK_KWH']


merged_ds.head()
```

```python
merged_ds.to_csv("Summert_data.csv")
```
